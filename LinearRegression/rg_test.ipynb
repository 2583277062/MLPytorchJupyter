```python
import numpy as np
from torch.utils.data import Dataset
device = 'cuda' if torch.cuda.is_available() else 'cpu'
class Covid19(Dataset):
    def __init__(self,trainset,test_set,mode='train',label_name='tested_positive.2'):
        
        self.mode =  mode
        target = trainset.loc[:,label_name]
        data = trainset.copy()
        if label_name in data.columns:
            data.pop(label_name)
        
        if mode == 'train':
            indices = [i for i in range(len(data)) if i%10 !=0]
            self.target = np.array(target)[indices]
            self.data = np.array(data)[indices,:]
        if mode == 'dev':
            indices = [i for i in range(len(data)) if i%10 ==0]
            self.target = np.array(target)[indices]
            self.data = np.array(data)[indices,:]
        if mode == 'test':
            self.data = np.array(test_set)
            self.target=None
            
        self.data = torch.FloatTensor(self.data)
        if not self.target is None:
            self.target = torch.FloatTensor(self.target)
            
        self.data = (self.data - self.data.mean(dim=0,keepdim=True))/self.data.std(dim=0,keepdim=True)
        self.dim = self.data.shape[1]
        
    def __getitem__(self,index):
        if self.mode in ['train','dev']:
            return self.data[index],self.target[index]
        else:
            return self.data[index]
    def __len__(self):
        return len(self.data)
        
    
Covid19(trainset,test_set,mode='train')
```

```python
from torch.utils.data import DataLoader
def pre_dataloader(trainset,test_set,mode,batch_size,n_jobs=0):
    dataset = Covid19(trainset,test_set,mode)
    dataloader = DataLoader(dataset,batch_size,shuffle=True,drop_last=False,num_workers=n_jobs,pin_memory=True)
    return dataloader
```

```python
traindata = pre_dataloader(trainset,test_set,'train',64)
```

```python
next(iter(traindata))
```

```python
import torch.nn as nn
class rgModel(nn.Module):
    def __init__(self, input_dim,output_dim):
        super(rgModel,self).__init__()
        self.layer1 = nn.Linear(input_dim,256)
        self.layer2 = nn.Linear(256, 16)
        self.layer3 = nn.Linear(16,output_dim)
        
        self.criterion = nn.MSELoss()
        
    def forward(self,x):
        x = self.layer1(x)
        nn.ReLU(x)
        x = self.layer2(x)
#         nn.ReLU(x)
        x = self.layer3(x)
        return x
    def cal_loss(self,pred,target):
        return self.criterion(pred,target)
```

```python
model = rgModel(traindata.dataset.dim,1).to(device)
```

```python
def train(traindata,model):
    optm = torch.optim.Adam(model.parameters(),lr=0.001)
    for epoch in range(1,25):
        model.train()
        for X, y in traindata:
            optm.zero_grad()
            X, y = X.to(device),y.to(device)
            pred = model(X)
#             print(pred)
            loss = model.cal_loss(pred,y)
            loss.backward()
            optm.step()
            print(loss.detach().cpu().item())
            
train(traindata,model)
```
